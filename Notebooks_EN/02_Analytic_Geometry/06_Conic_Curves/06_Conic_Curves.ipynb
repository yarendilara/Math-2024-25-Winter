### Conic Sections: Key Concepts and Formulas

Conic sections refer to the curves obtained by intersecting a plane with a cone. The basic types of conic sections are the **circle**, **ellipse**, **parabola**, and **hyperbola**. These curves are defined by specific equations, and understanding how to derive and manipulate these equations is crucial in geometry and calculus.

---

#### **Circle**
A **circle** is the set of all points in a plane that are equidistant from a fixed point, called the center.

- **Standard Equation of a Circle:**
  The standard equation of a circle with center \((h, k)\) and radius \(r\) is:  
  $$ (x - h)^2 + (y - k)^2 = r^2 $$

---

#### **Parabola**
A **parabola** is the set of all points equidistant from a fixed point, called the focus, and a fixed line, called the directrix.

- **Standard Equation of a Parabola:**
  The standard equation of a parabola opening vertically (up or down) is:  
  $$ y - k = a(x - h)^2 $$  
  The standard equation of a parabola opening horizontally (left or right) is:  
  $$ x - h = a(y - k)^2 $$  
  Where \((h, k)\) is the vertex and \(a\) is a constant that affects the width of the parabola.

---

#### **Ellipse**
An **ellipse** is the set of all points for which the sum of the distances to two fixed points, called foci, is constant.

- **Standard Equation of an Ellipse:**
  The standard equation of an ellipse centered at the origin is:  
  $$ \frac{x^2}{a^2} + \frac{y^2}{b^2} = 1 $$  
  Where \(a\) is the length of the semi-major axis and \(b\) is the length of the semi-minor axis.  
  If the center is at \((h, k)\), the equation becomes:  
  $$ \frac{(x - h)^2}{a^2} + \frac{(y - k)^2}{b^2} = 1 $$

---

#### **Hyperbola**
A **hyperbola** is the set of all points where the absolute difference of the distances to two fixed points (the foci) is constant.

- **Standard Equation of a Hyperbola:**
  The standard equation of a hyperbola centered at the origin is:  
  $$ \frac{x^2}{a^2} - \frac{y^2}{b^2} = 1 $$  
  Where \(a\) is the distance from the center to each vertex, and \(b\) is related to the asymptotes.  
  If the center is at \((h, k)\), the equation becomes:  
  $$ \frac{(x - h)^2}{a^2} - \frac{(y - k)^2}{b^2} = 1 $$

---

#### **Tangents to Conic Sections**
The tangent line to a conic section at a given point is the line that touches the curve at that point without crossing it.

- **Equation of the Tangent to a Circle:**
  The equation of the tangent to a circle \( (x - h)^2 + (y - k)^2 = r^2 \) at the point \((x_1, y_1)\) on the circle is:  
  $$ (x_1 - h)(x - h) + (y_1 - k)(y - k) = r^2 $$

---

$$Q 1$$

### Find the equation of a circle with center at point \( A(1,2) \) and radius \( r = 3 \)

The general equation of a circle with center \( (h, k) \) and radius \( r \) is:

$$ (x - h)^2 + (y - k)^2 = r^2 $$

Given:
- Center \( A(1, 2) \), so \( h = 1 \) and \( k = 2 \).
- Radius \( r = 3 \).

Substituting these values into the equation:

$$ (x - 1)^2 + (y - 2)^2 = 3^2 $$  
$$ (x - 1)^2 + (y - 2)^2 = 9 $$

Thus, the equation of the circle is:

$$ (x - 1)^2 + (y - 2)^2 = 9 $$


$$Q 2$$

### Find the equation of a parabola intersecting the \( Ox \) axis at points \( x = 2 \), \( x = 4 \), and passing through point \( y(3) = 1 \)

The general form of the equation of a parabola is:

$$ y = a(x - h)^2 + k $$

Given that the parabola intersects the \( Ox \) axis at \( x = 2 \) and \( x = 4 \), the equation can be written as:

$$ y = a(x - 2)(x - 4) $$

We use the point \( (3, 1) \) to find the value of \( a \):

Substitute \( x = 3 \) and \( y = 1 \):

$$ 1 = a(3 - 2)(3 - 4) $$  
$$ 1 = a(1)(-1) $$  
$$ 1 = -a $$  
Thus, \( a = -1 \).

Therefore, the equation of the parabola is:

$$ y = -(x - 2)(x - 4) $$

Expanding:

$$ y = -x^2 + 6x - 8 $$

Thus, the equation of the parabola is:

$$ y = -x^2 + 6x - 8 $$


$$Q 3$$

### Find the center of the ellipse with the equation \( x^2 + 4y^2 - 4x - 16y + 16 = 0 \)

We complete the square for both \( x \) and \( y \):

The given equation is:

$$ x^2 + 4y^2 - 4x - 16y + 16 = 0 $$

Group the \( x \)-terms and \( y \)-terms:

$$ (x^2 - 4x) + 4(y^2 - 4y) + 16 = 0 $$

___
### Complete the square:

- For \( x^2 - 4x \), add and subtract \( 4 \):

$$ x^2 - 4x = (x^2 - 4x + 4) - 4 = (x - 2)^2 - 4 $$

- For \( y^2 - 4y \), factor out \( 4 \), then complete the square for \( y^2 - 4y \) (add \( 4 \)):

$$ y^2 - 4y = 4\left(y^2 - y\right) $$

Now, complete the square inside the parentheses:

$$ y^2 - y = (y - \frac{1}{2})^2 - \frac{1}{4} $$

Thus, we have:

$$ 4\left(y^2 - y\right) = 4\left( (y - \frac{1}{2})^2 - \frac{1}{4} \right) = 4(y - \frac{1}{2})^2 - 1 $$


Thus:

$$ (x^2 - 4x + 4) + 4(y^2 - 4y + 4) + 16 - 4 - 16 = 0 $$  
$$ (x - 2)^2 + 4(y - 2)^2 = 4 $$

Now divide by 4:

$$ \frac{(x - 2)^2}{4} + \frac{(y - 2)^2}{1} = 1 $$

The center of the ellipse is at \( (2, 2) \).

The distance from a point \( (x_1, y_1) \) to a line \( Ax + By + C = 0 \) is given by:

$$ d = \frac{|Ax_1 + By_1 + C|}{\sqrt{A^2 + B^2}} $$

For the line \( y = mx - 5 \), we rewrite it as:

$$ mx - y - 5 = 0 $$

The center of the circle is at \( (0, 0) \). Thus, the distance from the center to the line is:

$$ d = \frac{|m(0) - 1(0) - 5|}{\sqrt{m^2 + 1}} $$  
$$ d = \frac{| - 5 |}{\sqrt{m^2 + 1}} $$  
$$ d = \frac{5}{\sqrt{m^2 + 1}} $$

For tangency, the distance must equal the radius (1):

$$ \frac{5}{\sqrt{m^2 + 1}} = 1 $$

Solve for \( m \):

$$ 5 = \sqrt{m^2 + 1} $$  
$$ 25 = m^2 + 1 $$  
$$ m^2 = 24 $$  
$$ m = \sqrt{24} $$  
$$ m = 2\sqrt{6} $$

### Find the intersection points of the hyperbola \( x^2 - y^2 = 1 \) with the ellipse's line \( x^2 + 4y^2 = 6 \)

We solve the system of equations:

$$ 1. \quad x^2 - y^2 = 1 $$  
$$ 2. \quad x^2 + 4y^2 = 6 $$

From the first equation, solve for \( x^2 \):

$$ x^2 = y^2 + 1 $$

Substitute into the second equation:

$$ (y^2 + 1) + 4y^2 = 6 $$  
$$ y^2 + 1 + 4y^2 = 6 $$  
$$ 5y^2 + 1 = 6 $$  
$$ 5y^2 = 5 $$  
$$ y^2 = 1 $$  
$$ y = \pm 1 $$

Substitute \( y = 1 \) and \( y = -1 \) into \( x^2 = y^2 + 1 \):

For \( y = 1 \):  
$$ x^2 = 1 + 1 = 2 $$  
$$ x = \pm \sqrt{2} $$

For \( y = -1 \):  
$$ x^2 = 1 + 1 = 2 $$  
$$ x = \pm \sqrt{2} $$

Thus, the intersection points are $$ (\sqrt{2}, 1) $$, $$ (-\sqrt{2}, 1) $$, $$ (\sqrt{2}, -1) $$, and $$ (-\sqrt{2}, -1) $$.


$$Q 6$$

### For the given hyperbola \( x^2 - y^2 = 1 \), find the distance between its branches


For the equation \( x^2 - y^2 = 1 \), we compare it with the standard form:

$$ \frac{x^2}{1} - \frac{y^2}{1} = 1 $$

Here, \( a^2 = 1 \) and \( b^2 = 1 \), so \( b = 1 \).

Thus, the distance between the branches is:

$$ 2b = 2 \times 1 = 2 $$
