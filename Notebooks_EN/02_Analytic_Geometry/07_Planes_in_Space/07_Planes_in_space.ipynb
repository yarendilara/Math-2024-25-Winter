# **Equations of Planes in Space**

Planes are fundamental geometric objects in three-dimensional space. The equation of a plane is typically written in the general form:

$$
Ax + By + Cz + D = 0
$$

where $A$, $B$, and $C$ are the coefficients representing the direction ratios of the normal vector to the plane, and $D$ is a constant. Several techniques are used to find the equation of a plane depending on the given information.

## **1. Equation of a Plane through Three Points**

Given three points $A(x_1, y_1, z_1)$, $B(x_2, y_2, z_2)$, and $C(x_3, y_3, z_3)$, the equation of the plane passing through these points can be found using the following steps:

- First, find the vectors $\vec{AB}$ and $\vec{AC}$:

$$
\vec{AB} = (x_2 - x_1, y_2 - y_1, z_2 - z_1)
$$

$$
\vec{AC} = (x_3 - x_1, y_3 - y_1, z_3 - z_1)
$$

- Next, find the cross product $\vec{AB} \times \vec{AC}$, which gives the normal vector $\vec{n}$ to the plane:

$$
\vec{n} = \vec{AB} \times \vec{AC}
$$

- Finally, use the point $A(x_1, y_1, z_1)$ and the normal vector $\vec{n} = (A, B, C)$ in the plane equation:

$$
A(x - x_1) + B(y - y_1) + C(z - z_1) = 0
$$

## **2. Equation of a Plane Parallel to Another Plane**

If a plane passes through a point $A(x_0, y_0, z_0)$ and is parallel to a given plane $Ax + By + Cz + D = 0$, the equation of the new plane will have the same normal vector $(A, B, C)$. Thus, the equation becomes:

$$
A(x - x_0) + B(y - y_0) + C(z - z_0) = 0
$$

where $A(x_0, y_0, z_0)$ is the point through which the plane passes.

## **3. Equation of a Plane Perpendicular to a Vector**

If a plane is perpendicular to a given normal vector $\vec{n} = [A, B, C]$ and passes through point $A(x_0, y_0, z_0)$, the equation of the plane is:

$$
A(x - x_0) + B(y - y_0) + C(z - z_0) = 0
$$

This equation holds since the vector $\vec{n} = [A, B, C]$ is perpendicular to the plane.

## **4. Line of Intersection of Two Planes**

Given two planes with equations $P_1: A_1x + B_1y + C_1z + D_1 = 0$ and $P_2: A_2x + B_2y + C_2z + D_2 = 0$, the line of intersection can be found as follows:

- Solve the system of equations for $x$, $y$, and $z$. One approach is to eliminate one of the variables (say, $x$ or $y$) using the equations of the planes.
- The direction of the line is given by the cross product of the normal vectors of the two planes $\vec{n_1} = [A_1, B_1, C_1]$ and $\vec{n_2} = [A_2, B_2, C_2]$, which will be parallel to the line of intersection.

$$
\vec{v} = \vec{n_1} \times \vec{n_2}
$$

## **5. Equation of a Plane Parallel to Two Vectors**

Given a point $A(x_0, y_0, z_0)$ and two vectors $\vec{v_1} = [a_1, b_1, c_1]$ and $\vec{v_2} = [a_2, b_2, c_2]$ that are parallel to the plane, the normal vector to the plane can be found as the cross product:

$$
\vec{n} = \vec{v_1} \times \vec{v_2}
$$

The equation of the plane is then:

$$
n_1(x - x_0) + n_2(y - y_0) + n_3(z - z_0) = 0
$$

where $\vec{n} = [n_1, n_2, n_3]$ is the normal vector.

## **6. Parallel and Perpendicular Planes**

Given a plane equation $2x + 3y + 4z = 5$, any plane parallel to it will have the same normal vector $\vec{n} = [2, 3, 4]$, but a different constant $D$. For example:

$$
2x + 3y + 4z = 10 \quad \text{(parallel plane)}
$$

To find a perpendicular plane, we can use the fact that the normal vector of the perpendicular plane will have a different direction, but the relation between the two planes can be explored through dot product or cross product operations.

## **7. Distance from a Point to a Plane**

Given a plane $Ax + By + Cz + D = 0$ and a point $A(x_0, y_0, z_0)$, the distance $d$ from the point to the plane is given by:

$$
d = \frac{|Ax_0 + By_0 + Cz_0 + D|}{\sqrt{A^2 + B^2 + C^2}}
$$

This formula gives the shortest distance from the point to the plane.

## **8. Equation of the Plane Intersecting the Coordinate Axes**

If a plane intersects the coordinate axes at points $A(x_1, 0, 0)$, $B(0, y_2, 0)$, and $C(0, 0, z_3)$, the equation of the plane can be written in intercept form:

$$
\frac{x}{x_1} + \frac{y}{y_2} + \frac{z}{z_3} = 1
$$

where $x_1$, $y_2$, $z_3$ are the intercepts on the $x$-, $y$-, and $z$-axes, respectively.

## **9. Angle Between Two Planes**

The angle $\theta$ between two planes is given by the formula:

$$
\cos(\theta) = \frac{|\vec{n_1} \cdot \vec{n_2}|}{|\vec{n_1}| |\vec{n_2}|}
$$

where $\vec{n_1}$ and $\vec{n_2}$ are the normal vectors to the planes.

For the planes $x + y + z = 1$ and $x = 0$ (the $yz$-plane), the normal vectors are $\vec{n_1} = [1, 1, 1]$ and $\vec{n_2} = [1, 0, 0]$, respectively. The angle between them can be calculated using the above formula.

## **10. Perpendicular Vector to a Plane**

For the plane $x + y + z = 1$, the normal vector is simply $\vec{n} = [1, 1, 1]$, which is perpendicular to the plane.

$$

$$

2. **Find the normal vector to the plane:**

The normal vector is given by the cross product of $\vec{AB}$ and $\vec{AC}$:

$$
\vec{n} = \vec{AB} \times \vec{AC} = \begin{vmatrix} \hat{i} & \hat{j} & \hat{k} \\ 2 & 2 & 2 \\ 1 & -1 & 1 \end{vmatrix}
$$

$$
\vec{n} = \hat{i} \left( \begin{vmatrix} 2 & 2 \\ -1 & 1 \end{vmatrix} \right) - \hat{j} \left( \begin{vmatrix} 2 & 2 \\ 1 & 1 \end{vmatrix} \right) + \hat{k} \left( \begin{vmatrix} 2 & 2 \\ 1 & -1 \end{vmatrix} \right)
$$

$$
\vec{n} = \hat{i} (2 \times 1 - 2 \times -1) - \hat{j} (2 \times 1 - 2 \times 1) + \hat{k} (2 \times -1 - 2 \times 1)
$$

$$
\vec{n} = \hat{i} (2 + 2) - \hat{j} (2 - 2) + \hat{k} (-2 - 2)
$$

$$
\vec{n} = \hat{i} (4) - \hat{j} (0) + \hat{k} (-4)
$$

$$
\vec{n} = (4, 0, -4)
$$

$$

$$

$$

$$

$$

$$
\vec{d} = \hat{i} \begin{vmatrix} 3 & 4 \\ 4 & 2 \end{vmatrix} - \hat{j} \begin{vmatrix} 2 & 4 \\ 3 & 2 \end{vmatrix} + \hat{k} \begin{vmatrix} 2 & 3 \\ 3 & 4 \end{vmatrix}

$$
\vec{d} = \hat{i} (3 \times 2 - 4 \times 4) - \hat{j} (2 \times 2 - 4 \times 3) + \hat{k} (2 \times 4 - 3 \times 3)

$$
\vec{d} = \hat{i} (6 - 16) - \hat{j} (4 - 12) + \hat{k} (8 - 9)
$$

$$

$$
\vec{d} = (-10, 8, -1)

To find a point on the line, solve the system of equations:

$$
2x + 3y + 4z = 5
$$

$$
3x + 4y + 2z = 6
$$

Set $z = 0$, and solve the system:

$$
2x + 3y = 5
$$

$$

$$

$$

$$

The distance \( d \) from a point \( (x_1, y_1, z_1) \) to a plane \( ax + by + cz + d = 0 \) is:

$$
d = \frac{|a x_1 + b y_1 + c z_1 + d|}{\sqrt{a^2 + b^2 + c^2}}
$$

$$

$$
d = \frac{|2 + 6 + 12 - 5|}{\sqrt{4 + 9 + 16}}
$$

$$
