## Tasks related to equations of planes in space

**Visualize each task.** - Python, GeoGebra, Wolfram Alpha, ...

* The plane passes through points $A(1, 2, 3)$, $B(3, 4, 5)$, and $C(2, 1, 4)$. Find the equation of the plane.
* The plane passes through point $A(1, 2, 3)$ and is parallel to the plane $2x + 3y + 4z = 5$. Find the equation of the plane.
* The plane passes through point $A(1, 2, 3)$ and is perpendicular to the normal vector $\vec{n} = [2, 3, 4]$. Find the equation of the plane.
* We have two planes $2x + 3y + 4z = 5$ and $3x + 4y + 2z = 6$. Find the line of intersection of these planes.
* Write the equation of the plane passing through point $A(1, 2, 3)$ and parallel to vectors $\vec{v_1} = [1, 0, 1]$ and $\vec{v_2} = [0, 1, -1]$.
* We have the plane $2x + 3y + 4z = 5$. Find an example of a plane parallel and perpendicular to it.
* We have the plane $2x + 3y + 4z = 5$ and point $A(1, 2, 3)$. Find the distance from point $A$ to this plane.
* The plane intersects the coordinate axes at points $A(2, 0, 0)$, $B(0, 3, 0)$, and $C(0, 0, 4)$. Find the equation of the plane.
* Calculate the angle between the plane $x + y + z = 1$ and the plane $x = 0$ (i.e., the $yz$ plane).
* Find the vector perpendicular to the plane $x + y + z = 1$.

$$Q1$$

1. **Find two vectors on the plane:**

$$
\vec{AB} = B - A = (3-1, 4-2, 5-3) = (2, 2, 2)
$$

$$
\vec{AC} = C - A = (2-1, 1-2, 4-3) = (1, -1, 1)
$$

2. **Find the normal vector to the plane:**

The normal vector is given by the cross product of $\vec{AB}$ and $\vec{AC}$:

$$
\vec{n} = \vec{AB} \times \vec{AC} = \begin{vmatrix} \hat{i} & \hat{j} & \hat{k} \\ 2 & 2 & 2 \\ 1 & -1 & 1 \end{vmatrix}
$$

$$
\vec{n} = \hat{i} \left( \begin{vmatrix} 2 & 2 \\ -1 & 1 \end{vmatrix} \right) - \hat{j} \left( \begin{vmatrix} 2 & 2 \\ 1 & 1 \end{vmatrix} \right) + \hat{k} \left( \begin{vmatrix} 2 & 2 \\ 1 & -1 \end{vmatrix} \right)
$$

$$
\vec{n} = \hat{i} (2 \times 1 - 2 \times -1) - \hat{j} (2 \times 1 - 2 \times 1) + \hat{k} (2 \times -1 - 2 \times 1)
$$

$$
\vec{n} = \hat{i} (2 + 2) - \hat{j} (2 - 2) + \hat{k} (-2 - 2)
$$

$$
\vec{n} = \hat{i} (4) - \hat{j} (0) + \hat{k} (-4)
$$

$$
\vec{n} = (4, 0, -4)
$$

3. **Equation of the plane:**

Using the point $A(1, 2, 3)$ and the normal vector $(4, 0, -4)$, the equation of the plane is:

$$
4(x - 1) + 0(y - 2) - 4(z - 3) = 0
$$

Simplifying:

$$
4x - 4 + 0 - 4z + 12 = 0
$$

$$
4x - 4z + 8 = 0
$$

$$
x - z + 2 = 0
$$

The equation of the plane is:

$$
x - z + 2 = 0
$$

$$Q5$$

1. **Find the normal vectors to the planes:**

The normal vector to the first plane is $(2, 3, 4)$ and to the second plane is $(3, 4, 2)$.

2. **Find the direction of the line:**

The direction of the line is the cross product of the normal vectors:

$$
\vec{d} = \vec{n_1} \times \vec{n_2} = \begin{vmatrix} \hat{i} & \hat{j} & \hat{k} \\ 2 & 3 & 4 \\ 3 & 4 & 2 \end{vmatrix}
$$

$$
\vec{d} = \hat{i} \begin{vmatrix} 3 & 4 \\ 4 & 2 \end{vmatrix} - \hat{j} \begin{vmatrix} 2 & 4 \\ 3 & 2 \end{vmatrix} + \hat{k} \begin{vmatrix} 2 & 3 \\ 3 & 4 \end{vmatrix}
$$

$$
\vec{d} = \hat{i} (3 \times 2 - 4 \times 4) - \hat{j} (2 \times 2 - 4 \times 3) + \hat{k} (2 \times 4 - 3 \times 3)
$$

$$
\vec{d} = \hat{i} (6 - 16) - \hat{j} (4 - 12) + \hat{k} (8 - 9)
$$

$$
\vec{d} = \hat{i} (-10) - \hat{j} (-8) + \hat{k} (-1)
$$

$$
\vec{d} = (-10, 8, -1)
$$

3. **Find a point on the line:**

To find a point on the line, solve the system of equations:

$$
2x + 3y + 4z = 5
$$

$$
3x + 4y + 2z = 6
$$

Set $z = 0$, and solve the system:

$$
2x + 3y = 5
$$

$$
3x + 4y = 6
$$

Multiply the first equation by 3 and the second equation by 2:

$$
6x + 9y = 15
$$

$$
6x + 8y = 12
$$

Subtract the second equation from the first:

$$
y = 3
$$

Substitute $y = 3$ into $2x + 3y = 5$:

$$
2x + 9 = 5
$$

$$
2x = -4
$$

$$
x = -2
$$

Thus, the point is $(-2, 3, 0)$.

4. **The parametric equations of the line:**

The parametric equations of the line are:

$$
x = -2 - 10t
$$

$$
y = 3 + 8t
$$

$$
z = -t
$$

Thus, the line of intersection is:

$$
(x, y, z) = (-2 - 10t, 3 + 8t, -t)
$$


$$Q5$$



1. **Find the normal vector to the plane:**

The normal vector to the plane is the cross product of \( \vec{v_1} \) and \( \vec{v_2} \):

$$
\vec{n} = \vec{v_1} \times \vec{v_2} = \begin{vmatrix} \hat{i} & \hat{j} & \hat{k} \\ 1 & 0 & 1 \\ 0 & 1 & -1 \end{vmatrix}
$$

$$
\vec{n} = \hat{i} \begin{vmatrix} 0 & 1 \\ 1 & -1 \end{vmatrix} - \hat{j} \begin{vmatrix} 1 & 1 \\ 0 & -1 \end{vmatrix} + \hat{k} \begin{vmatrix} 1 & 0 \\ 0 & 1 \end{vmatrix}
$$

$$
\vec{n} = \hat{i} (0 \times -1 - 1 \times 1) - \hat{j} (1 \times -1 - 1 \times 0) + \hat{k} (1 \times 1 - 0 \times 0)
$$

$$
\vec{n} = \hat{i} (-1) - \hat{j} (-1) + \hat{k} (1)
$$

$$
\vec{n} = (-1, 1, 1)
$$

2. **Equation of the plane:**

Using the point \( A(1, 2, 3) \) and the normal vector \( (-1, 1, 1) \), the equation of the plane is:

$$
-1(x - 1) + 1(y - 2) + 1(z - 3) = 0
$$

Simplify:

$$
-x + 1 + y - 2 + z - 3 = 0
$$

$$
-x + y + z - 4 = 0
$$

The equation of the plane is:

$$
x - y - z + 4 = 0
$$


$$Q6$$

1. **Parallel plane:**

A plane parallel to \( 2x + 3y + 4z = 5 \) has the same normal vector \( (2, 3, 4) \). An example is:

$$
2x + 3y + 4z = 10
$$

2. **Perpendicular plane:**

A plane perpendicular to \( 2x + 3y + 4z = 5 \) must have a normal vector orthogonal to \( (2, 3, 4) \). Let the normal vector of the new plane be \( (a, b, c) \). Then:

$$
2a + 3b + 4c = 0
$$

Choosing \( a = 4, b = -2, c = 0 \) satisfies this equation. The equation of the plane is:

$$
4x - 2y = k
$$

For simplicity, let \( k = 0 \):

$$
4x - 2y = 0
$$


$$Q7$$

1. **Distance formula:**

The distance \( d \) from a point \( (x_1, y_1, z_1) \) to a plane \( ax + by + cz + d = 0 \) is:

$$
d = \frac{|a x_1 + b y_1 + c z_1 + d|}{\sqrt{a^2 + b^2 + c^2}}
$$

2. **Substitute the values:**

For the plane \( 2x + 3y + 4z - 5 = 0 \), \( a = 2, b = 3, c = 4, d = -5 \), and for the point \( A(1, 2, 3) \):

$$
d = \frac{|2(1) + 3(2) + 4(3) - 5|}{\sqrt{2^2 + 3^2 + 4^2}}
$$

$$
d = \frac{|2 + 6 + 12 - 5|}{\sqrt{4 + 9 + 16}}
$$

$$
d = \frac{|15|}{\sqrt{29}}
$$

$$
d = \frac{15}{\sqrt{29}}
$$

The distance is:

$$
\frac{15}{\sqrt{29}}
$$
