# **Integrals**


Integration is a fundamental concept in calculus that represents the accumulation of quantities and the inverse process of differentiation. It is widely used in mathematics, physics, engineering, and economics to compute areas, volumes, and solve differential equations.

### **Types of Integrals**
There are two main types of integrals:
1. **Indefinite Integrals**: These represent a family of functions whose derivative gives the original function. The result of an indefinite integral includes an arbitrary constant $C$, since differentiation removes constant terms.
2. **Definite Integrals**: These compute the accumulated quantity over an interval $[a, b]$. They are expressed as:
   $$
   \int_a^b f(x) \,dx
   $$
   and are evaluated using the Fundamental Theorem of Calculus.

## **Basic Integration Rules**

### **1. Power Rule**
For any real number $n \neq -1$:
$$
\int x^n \,dx = \frac{x^{n+1}}{n+1} + C
$$
This rule extends differentiation in reverse, increasing the exponent instead of decreasing it.

### **2. Constant Rule**
If $c$ is a constant:
$$
\int c \,dx = cx + C
$$
This follows directly from the derivative of a linear function.

### **3. Sum and Difference Rule**
Integration distributes over addition and subtraction:
$$
\int [f(x) + g(x)] \,dx = \int f(x) \,dx + \int g(x) \,dx
$$
This allows the integral of a sum to be computed as the sum of individual integrals.

### **4. Exponential and Logarithmic Functions**
- The integral of an exponential function is:
  $$
  \int e^x \,dx = e^x + C
  $$
- The integral of the natural logarithm follows:
  $$
  \int \frac{1}{x} \,dx = \ln |x| + C
  $$

### **5. Trigonometric Integrals**
Basic trigonometric functions integrate as follows:
- $$\int \sin x \,dx = -\cos x + C$$
- $$\int \cos x \,dx = \sin x + C$$
- $$\int \sec^2 x \,dx = \tan x + C$$

### **6. Integration by Substitution**
Used when an integral contains a composite function. If $u = g(x)$, then:
$$
\int f(g(x)) g'(x) \,dx = \int f(u) \,du
$$
This method simplifies complex integrals by changing the variable.

### **7. Integration by Parts**
Based on the product rule of differentiation:
$$
\int u \,dv = uv - \int v \,du
$$
This method is useful for integrating products of functions such as $x e^x$ or $x \ln x$.

### **8. Definite Integrals and the Fundamental Theorem of Calculus**
The Fundamental Theorem of Calculus connects differentiation and integration:
$$
\int_a^b f(x) \,dx = F(b) - F(a)
$$
where $F(x)$ is an antiderivative of $f(x)$. This theorem allows us to evaluate definite integrals without needing summation approximations.

## **Applications of Integration**
Integration is widely used in:
- Calculating **areas** under curves
- Determining **volumes** of solids of revolution using the disk and shell methods
- Computing **work** and **energy** in physics
- Solving **differential equations**
- Finding **arc lengths** and **surface areas**

Mastering integration requires a deep understanding of these fundamental rules and techniques. Advanced topics include improper integrals, numerical integration, and integral transforms such as Laplace and Fourier transforms.

By applying these principles effectively, one can solve a wide range of mathematical and real-world problems with precision and elegance.

### **(a) $\int 1 \,dx$**
Using the basic rule of integration:
$$
\int c \,dx = c x + C

where $c = 1$, we get:

### **(b) $\int (x^2 + 2) \,dx$**
Using the sum rule:
$$
\int [f(x) + g(x)] \,dx = \int f(x) \,dx + \int g(x) \,dx
$$

Breaking into two integrals:
$$
\int x^2 \,dx + \int 2 \,dx
$$

#### **First Integral: $\int x^2 \,dx$**
Using the power rule:
$$
\int x^n \,dx = \frac{x^{n+1}}{n+1} + C
$$
where $n = 2$:
$$
\int x^2 \,dx = \frac{x^3}{3}
$$

#### **Second Integral: $\int 2 \,dx$**
Since $2$ is a constant:
$$
\int 2 \,dx = 2x
$$

#### **Final Answer**
$$
\frac{x^3}{3} + 2x + C
$$

### **(c) $\int 2\sin(x) \,dx$**
Using:
$$
\int \sin(x) \,dx = -\cos(x) + C
$$
Since we have a constant $2$:
$$
2 \int \sin(x) \,dx = -2\cos(x) + C
$$

### **(d) $\int \frac{3}{x} \,dx$**
Using:
$$
\int \frac{1}{x} \,dx = \ln |x| + C
$$
Thus:
$$
\int \frac{3}{x} \,dx = 3 \ln |x| + C
$$

### **(e) $\int \frac{1}{x^2} \,dx$**
Rewriting $\frac{1}{x^2}$ as $x^{-2}$:
$$
\int x^{-2} \,dx = \frac{x^{-1}}{-1} = -\frac{1}{x} + C
$$

### **(f) $\int \left( \frac{1}{3}x^4 - 5 \right) \, dx$**
Using sum rule:
$$
\int \frac{1}{3}x^4 \,dx - \int 5 \,dx
$$

#### **First Integral**
$$
\frac{1}{3} \int x^4 \,dx = \frac{1}{3} \cdot \frac{x^5}{5} = \frac{x^5}{15}
$$

#### **Second Integral**
$$
\int 5 \,dx = 5x
$$

#### **Final Answer**
$$
\frac{x^5}{15} - 5x + C
$$

### **(g) $\int (\sin^2 x + \cos^2 x) \,dx$**

\sin^2 x + \cos^2 x = 1

\int 1 \,dx = x + C

### **(i) $\int \sqrt[3]{x} \,dx$**
Rewriting:
$$
\sqrt[3]{x} = x^{1/3}
$$
Using power rule:
$$
\int x^{1/3} \,dx = \frac{x^{4/3}}{4/3} = \frac{3}{4} x^{4/3} + C
$$

### **(j) $\int \sqrt{10x} \,dx$**
Rewriting:
$$
\sqrt{10x} = (10x)^{1/2}
$$
Factoring out 10:
$$
10^{1/2} \int x^{1/2} \,dx
$$
Using power rule:
$$
\int x^{1/2} \,dx = \frac{x^{3/2}}{3/2} = \frac{2}{3} x^{3/2}
$$
Thus:
$$
\sqrt{10} \cdot \frac{2}{3} x^{3/2} + C
$$

### **(k) $\int \cos\left(\frac{5}{2}x + 3\right) \,dx$**
Using substitution:
Let $ u = \frac{5}{2}x + 3 $, then $ du = \frac{5}{2} dx $:
$$
\int \cos(u) \cdot \frac{2}{5} \,du
$$
Since $ \int \cos u \,du = \sin u $:
$$
\frac{2}{5} \sin\left(\frac{5}{2}x + 3\right) + C
$$

### **(l) $\int \frac{\cos(\ln x)}{x} \,dx$**

$$
\int \cos(u) \,du
$$
$$
\sin u + C
$$
$$
\sin(\ln x) + C
