# Extremum
#(Maxima and Minima): Key Concepts and Methods

In calculus, extremum points refer to the values of a function where the function reaches its highest or lowest points. These are crucial in many optimization problems, such as maximizing profit, minimizing cost, or finding the optimal configuration of a system. To determine extremum points, we use the following methods:

---

## 1. Derivatives and Critical Points

To find the extrema (either maxima or minima) of a function, we first need to compute its **first derivative**. The first derivative of a function provides the rate of change or the slope at any point. The primary steps are:

- **First Derivative**: For a function $$ f(x) $$, its derivative $$ f'(x) $$ gives us the slope of the tangent at any point on the curve. The formula for the derivative depends on the type of function (e.g., power rule, product rule, quotient rule, chain rule).
  
- **Critical Points**: Critical points occur where the first derivative equals zero, i.e., when $$ f'(x) = 0 $$ or when the first derivative does not exist. These points are potential locations for maxima, minima, or points of inflection.

  To find the critical points, solve for $$ x $$ when $$ f'(x) = 0 $$.

---

## 2. Second Derivative Test

After finding the critical points, we use the **second derivative test** to determine whether the function has a local maximum, local minimum, or if it's an inflection point.

- **Second Derivative**: The second derivative, denoted as $$ f''(x) $$, tells us about the concavity of the function:
  - If $$ f''(x) > 0 $$ at a critical point, the function is concave up, and we have a **local minimum** at that point.
  - If $$ f''(x) < 0 $$ at a critical point, the function is concave down, and we have a **local maximum** at that point.
  - If $$ f''(x) = 0 $$, the test is inconclusive, and further analysis (e.g., higher-order derivatives or analyzing the function's behavior near the point) is needed.

---

## 3. Optimization Problems:

Optimization problems often involve finding the maximum or minimum of a given function subject to constraints (e.g., maximizing profit, minimizing cost, or finding optimal dimensions). These problems typically follow this approach:

- **Objective Function**: The function that we want to maximize or minimize. This function usually involves variables representing quantities such as cost, area, or profit.
  
- **Constraints**: The restrictions or conditions that must be satisfied in the problem. For example, if you're maximizing the area of a rectangle, you may have a constraint like the perimeter being fixed.

  If you have a constraint, you may use the **method of Lagrange multipliers** or substitute one equation into another to reduce the problem to one variable.

---

## 4. Applying to Different Types of Functions

- **Quadratic Functions**: For functions of the form $$ f(x) = ax^2 + bx + c $$, the critical point can be easily found using the first derivative. The function will always have either a maximum or minimum, and the second derivative helps us classify it as such.
  
- **Rational Functions**: For functions of the form $$ f(x) = \frac{g(x)}{h(x)} $$ (a ratio of two functions), we apply the **quotient rule** to differentiate. Then we solve $$ f'(x) = 0 $$ to find the critical points, and use the second derivative to classify them.

- **Trigonometric, Exponential, and Logarithmic Functions**: For these types of functions, we apply standard differentiation rules:
  - **Trigonometric Functions**: $$ \frac{d}{dx} (\sin(x)) = \cos(x), \quad \frac{d}{dx} (\cos(x)) = -\sin(x) $$
  - **Exponential Functions**: $$ \frac{d}{dx} (e^x) = e^x $$
  - **Logarithmic Functions**: $$ \frac{d}{dx} (\ln(x)) = \frac{1}{x} $$

---

## 5. Conclusion:

To summarize, solving extremum problems involves:
1. **Differentiating the function** to find its **first derivative**.
2. **Finding critical points** by solving $$ f'(x) = 0 $$.
3. **Using the second derivative test** to classify the critical points as maxima, minima, or points of inflection.
4. If dealing with constraints, applying methods like Lagrange multipliers to optimize the function subject to the given restrictions.

The combination of the first and second derivative tests, along with the appropriate differentiation rules, enables us to solve complex optimization problems in many real-world applications.

where u is the number of units sold. Find the number of units that maximize profit.

2. You have 10 meters of string, and you need to use it to enclose the largest possible rectangle. Find the dimensions of the rectangle.

3. Find the extremum of $$ f(x) = x^2 + 3x - 5 $$

4. Find the extremum of $$ f(x) = \frac{x^2 + 2x + 1}{x - 1} $$


### 1. The profit function is $$ P(u) = -2u^2 + 50u - 300 $$, where $$ u $$ is the number of units sold. Find the number of units that maximize profit.

To find the number of units that maximize profit, we need to determine the **extremum** (specifically the maximum) of the profit function. The first step in this process is to calculate the **first derivative** of the profit function to find the critical points.

#### Step 1: Find the first derivative of the profit function.

The profit function is:

$$ P(u) = -2u^2 + 50u - 300 $$

We differentiate this function with respect to $$ u $$:

$$ \frac{dP(u)}{du} = \frac{d}{du} (-2u^2 + 50u - 300) $$

Apply the power rule of differentiation (the derivative of $$ u^n $$ is $$ n u^{n-1} $$):

$$ \frac{dP(u)}{du} = -4u + 50 $$

#### Step 2: Find the critical points.

To find the critical points, set the first derivative equal to zero:

$$ -4u + 50 = 0 $$

Solve for $$ u $$:

$$ -4u = -50 $$

$$ u = \frac{50}{4} $$

$$ u = 12.5 $$

So, the critical point occurs when $$ u = 12.5 $$.

#### Step 3: Determine whether this critical point is a maximum or minimum.

To determine whether this is a maximum or a minimum, we need to calculate the second derivative of the profit function and check its sign.

The second derivative is:

$$ \frac{d^2P(u)}{du^2} = \frac{d}{du}(-4u + 50) $$

$$ \frac{d^2P(u)}{du^2} = -4 $$

Since the second derivative is negative, the critical point at $$ u = 12.5 $$ is a **maximum**.

#### Conclusion:

The number of units that maximize the profit is $$ u = 12.5 $$.

### 2. You have 10 meters of string, and you need to use it to enclose the largest possible rectangle. Find the dimensions of the rectangle.

Here, we are trying to maximize the area of the rectangle with a fixed perimeter of 10 meters.

#### Step 1: Write down the perimeter equation.

The perimeter of a rectangle is given by:

$$ P = 2l + 2w $$

where $$ l $$ is the length and $$ w $$ is the width of the rectangle. We are told that the total perimeter is 10 meters, so:

$$ 2l + 2w = 10 $$

Simplifying:

$$ l + w = 5 $$

#### Step 2: Write down the area equation.

The area of the rectangle is:

$$ A = l \times w $$

Since we know that $$ l + w = 5 $$, we can express $$ l $$ in terms of $$ w $$:

$$ l = 5 - w $$

Now substitute this expression into the area formula:

$$ A = (5 - w) \times w $$

Expanding:

$$ A = 5w - w^2 $$

#### Step 3: Find the critical points.

To maximize the area, we need to find the critical points by differentiating the area function with respect to $$ w $$.

$$ \frac{dA}{dw} = \frac{d}{dw}(5w - w^2) $$

$$ \frac{dA}{dw} = 5 - 2w $$

Set the derivative equal to zero to find the critical point:

$$ 5 - 2w = 0 $$

$$ 2w = 5 $$

$$ w = \frac{5}{2} $$

So, the width that maximizes the area is $$ w = 2.5 $$ meters.

#### Step 4: Find the length.

Since $$ l + w = 5 $$ and $$ w = 2.5 $$, we have:

$$ l = 5 - 2.5 = 2.5 $$ meters.

#### Step 5: Verify that this is a maximum.

We can check the second derivative of the area function to verify that this is indeed a maximum:

$$ \frac{d^2A}{dw^2} = \frac{d}{dw}(5 - 2w) $$

$$ \frac{d^2A}{dw^2} = -2 $$

Since the second derivative is negative, the critical point at $$ w = 2.5 $$ is a maximum.

#### Conclusion:

The dimensions of the rectangle that maximize the area are 2.5 meters by 2.5 meters, which forms a **square**.

### 3. Find the extremum of $$ f(x) = x^2 + 3x - 5 $$.

#### Step 1: Find the first derivative.

To find the extremum, we first compute the first derivative of the function:

$$ f(x) = x^2 + 3x - 5 $$

Differentiating:

$$ f'(x) = 2x + 3 $$

#### Step 2: Find the critical points.

Set the first derivative equal to zero to find the critical points:

$$ 2x + 3 = 0 $$

Solve for $$ x $$:

$$ 2x = -3 $$

$$ x = -\frac{3}{2} $$

#### Step 3: Determine whether the critical point is a maximum or minimum.

To determine whether this is a maximum or minimum, we calculate the second derivative:

$$ f''(x) = \frac{d}{dx}(2x + 3) $$

$$ f''(x) = 2 $$

Since the second derivative is positive, the critical point at $$ x = -\frac{3}{2} $$ is a **minimum**.

#### Conclusion:

The function $$ f(x) = x^2 + 3x - 5 $$ has a minimum at $$ x = -\frac{3}{2} $$.

### 4. Find the extremum of $$ f(x) = \frac{x^2 + 2x + 1}{x - 1} $$.

#### Step 1: Find the first derivative.

We need to find the first derivative of the function using the **quotient rule**. The quotient rule states that if $$ f(x) = \frac{g(x)}{h(x)} $$, then:

$$ f'(x) = \frac{g'(x)h(x) - g(x)h'(x)}{(h(x))^2} $$

Let:

$$ g(x) = x^2 + 2x + 1 $$

$$ h(x) = x - 1 $$

First, we compute the derivatives of $$ g(x) $$ and $$ h(x) $$:

$$ g'(x) = 2x + 2 $$

$$ h'(x) = 1 $$

Now, applying the quotient rule:

$$ f'(x) = \frac{(2x + 2)(x - 1) - (x^2 + 2x + 1)(1)}{(x - 1)^2} $$

Expanding the terms:

$$ f'(x) = \frac{(2x^2 - 2x + 2x - 2) - (x^2 + 2x + 1)}{(x - 1)^2} $$

Simplify:

$$ f'(x) = \frac{2x^2 - 2 - x^2 - 2x - 1}{(x - 1)^2} $$

$$ f'(x) = \frac{x^2 - 2x - 3}{(x - 1)^2} $$

#### Step 2: Find the critical points.

Set the first derivative equal to zero:

$$ \frac{x^2 - 2x - 3}{(x - 1)^2} = 0 $$

The numerator must be zero for the entire fraction to be zero:

$$ x^2 - 2x - 3 = 0 $$

Factor the quadratic equation:

$$ (x - 3)(x + 1) = 0 $$

Thus, the critical points are:

$$ x = 3 \quad \text{or} \quad x = -1 $$
