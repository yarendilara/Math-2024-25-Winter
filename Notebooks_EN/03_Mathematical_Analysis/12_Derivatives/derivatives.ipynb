# Derivatives: A Comprehensive Overview

## 1. **Understanding Derivatives**

The derivative of a function gives us information about the function's rate of change at any given point. In simple terms, the derivative tells us how a function behaves locally, whether it is increasing, decreasing, or constant.

If we have a function **f(x)**, the derivative of **f(x)** at a specific point **x = a** is defined as:

$$
f'(a) = \lim_{\Delta x \to 0} \frac{f(a + \Delta x) - f(a)}{\Delta x}
$$

This expression represents the **instantaneous rate of change** of the function at the point **x = a**. The numerator represents the change in the function's value, and the denominator represents the change in **x**. As **Δx** gets closer to zero, the ratio becomes the derivative at that point.

---

## 2. **Derivatives of Polynomial Functions**

Consider the function **f(x) = 2x³ - 3x² + 8x - 9**. To find its derivative, we apply the **Power Rule**:

The **Power Rule** states that if you have a term of the form **axⁿ**, its derivative is:

$$
\frac{d}{dx} \left( ax^n \right) = n \cdot ax^{n-1}
$$

Let's break down the given function:

**f(x) = 2x³ - 3x² + 8x - 9**

- The derivative of **2x³** is **6x²** (using the Power Rule).
- The derivative of **-3x²** is **-6x**.
- The derivative of **8x** is **8**.
- The derivative of the constant **-9** is **0**.

Thus, the derivative of the function is:

$$
f'(x) = 6x^2 - 6x + 8
$$

This tells us how the function **f(x)** is changing at any point **x**.

---

## 3. **Trigonometric Derivatives**

Another important class of functions are **trigonometric functions**, such as sine and cosine. The derivatives of the most common trigonometric functions are:

- $$ \frac{d}{dx} (\sin(x)) = \cos(x) $$
- $$ \frac{d}{dx} (\cos(x)) = -\sin(x) $$

For example, let’s consider the function **f(x) = sin(x) + cos(x)**. To find its derivative, we apply the derivatives of sine and cosine:

$$
f'(x) = \frac{d}{dx} (\sin(x)) + \frac{d}{dx} (\cos(x)) = \cos(x) - \sin(x)
$$

---

## 4. **Product and Quotient Rules**

When we have functions that are products or quotients of other functions, we need to use the **Product Rule** and the **Quotient Rule**.

### **Product Rule**

For two functions **u(x)** and **v(x)**, the product rule states that:

$$
\frac{d}{dx} \left( u(x) \cdot v(x) \right) = u'(x) \cdot v(x) + u(x) \cdot v'(x)
$$

For example, consider **f(x) = x \cdot \sin(x)**. To find the derivative, we apply the Product Rule:

$$
f'(x) = \frac{d}{dx} (x) \cdot \sin(x) + x \cdot \frac{d}{dx} (\sin(x))
$$
$$
f'(x) = 1 \cdot \sin(x) + x \cdot \cos(x) = \sin(x) + x \cdot \cos(x)
$$

### **Quotient Rule**

For two functions **u(x)** and **v(x)**, the Quotient Rule is:

$$
\frac{d}{dx} \left( \frac{u(x)}{v(x)} \right) = \frac{v(x) \cdot u'(x) - u(x) \cdot v'(x)}{[v(x)]^2}
$$

For example, consider **f(x) = \frac{x}{x + 1}**. To find the derivative, we apply the Quotient Rule:

Let **u(x) = x** and **v(x) = x + 1**. The derivatives are **u'(x) = 1** and **v'(x) = 1**.

$$
f'(x) = \frac{(x + 1) \cdot 1 - x \cdot 1}{(x + 1)^2} = \frac{1}{(x + 1)^2}
$$

---

## 5. **Chain Rule**

The **Chain Rule** is used when we need to differentiate composite functions, that is, when one function is nested inside another.

If **y = g(f(x))**, the derivative is:

$$
\frac{dy}{dx} = \frac{dy}{df(x)} \cdot \frac{df(x)}{dx}
$$

For example, if we have **f(x) = sin(x²)**, this is a composite function where **g(u) = sin(u)** and **u = x²**. To differentiate this, we apply the Chain Rule:

$$
f'(x) = \cos(x²) \cdot \frac{d}{dx}(x²) = \cos(x²) \cdot 2x
$$

---

## 6. **Higher-Order Derivatives**

The **second derivative** is the derivative of the derivative of a function. It gives us information about the **concavity** of the function (whether it is curving upwards or downwards).

For example, if **f(x) = \cos(x)**:

1. The first derivative is: $$ f'(x) = -\sin(x) $$
2. The second derivative is: $$ f''(x) = -\cos(x) $$

Notice that the second derivative is the negative of the original function:

$$
f''(x) = -f(x)
$$

This shows that **f(x) = cos(x)** is a periodic function with alternating concavity.

---

## 7. **Applications of Derivatives**

### **Velocity and Acceleration**

In physics, the position of a particle as a function of time is denoted by **x(t)**. The **velocity** is the derivative of the position function with respect to time, and the **acceleration** is the derivative of the velocity.

If the position function is given by:

$$
x(t) = 3t^2 - 6t + 1
$$

To find the velocity, we take the derivative of the position:

$$
V(t) = x'(t) = 6t - 6
$$

To find the acceleration, we take the derivative of the velocity:

$$
a(t) = V'(t) = 6
$$

This tells us that the velocity at time **t = 2** is:

$$
V(2) = 6(2) - 6 = 6
$$

And the acceleration is constant:

$$
a(2) = 6
$$

---

## 8. **The Fundamental Theorem of Calculus**

The **Fundamental Theorem of Calculus** connects differentiation and integration. It states that if **F(x)** is an antiderivative of **f(x)**, then:

$$
\int_a^b f(x) \, dx = F(b) - F(a)
$$

In other words, the integral of a function over a range can be computed using its antiderivative. This theorem forms the foundation of calculus, allowing us to compute areas under curves and solve various physical problems.

$$ a(2) = 6 $$
