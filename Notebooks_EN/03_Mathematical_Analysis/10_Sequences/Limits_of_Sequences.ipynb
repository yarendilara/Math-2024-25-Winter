## Differential Equations and Limits of Sequences: A Comprehensive Guide

### 1. Introduction to Differential Equations

A differential equation is an equation that involves a function and its derivatives. These equations describe various physical, biological, and economic phenomena by modeling how a quantity changes with respect to another. The primary goal is to find a function that satisfies the given equation.

Differential equations are categorized based on their order, linearity, and whether they have constant or variable coefficients.

### 2. First-Order Differential Equations

A first-order differential equation involves only the first derivative of the function. These equations often appear in population dynamics, radioactive decay, and cooling processes. A common method for solving first-order differential equations is **separation of variables**.

#### 2.1. Separation of Variables
The method of separation of variables is applicable when the equation can be rewritten in the form:

$$ \frac{dy}{dx} = f(x) g(y) $$

By rewriting and integrating:

$$ \int \frac{dy}{g(y)} = \int f(x) dx $$

we obtain a solution in implicit form, which can often be solved explicitly for \( y \).

#### 2.2. Integrating Factor Method
For linear first-order equations of the form:

$$ \frac{dy}{dx} + P(x)y = Q(x) $$

we use an integrating factor:

$$ \mu(x) = e^{\int P(x)dx} $$

Multiplying the entire equation by \( \mu(x) \) allows it to be rewritten as an exact differential equation, which can then be integrated to find \( y(x) \).

### 3. Second-Order Differential Equations

A second-order differential equation involves the second derivative of the function. These equations frequently appear in mechanics, wave motion, and electrical circuits.

#### 3.1. Characteristic Equation Approach
For linear equations with constant coefficients:

$$ a y'' + b y' + c y = 0 $$

we assume a solution of the form:

$$ y = e^{rx} $$

Substituting into the equation gives a characteristic equation:

$$ ar^2 + br + c = 0 $$

Solving for \( r \) provides the general solution, which can involve real or complex exponentials depending on the nature of the roots.

#### 3.2. Homogeneous and Particular Solutions
For inhomogeneous equations:

$$ y'' + ay' + by = f(x) $$

we find the **complementary solution** (solving the homogeneous part) and a **particular solution** using methods like **undetermined coefficients** or **variation of parameters**.

### 4. Second-Order Partial Differential Equations

Partial differential equations (PDEs) involve multiple independent variables. A fundamental PDE is the **wave equation**:

$$ \frac{\partial^2 u}{\partial t^2} - v^2 \frac{\partial^2 u}{\partial x^2} = 0 $$

Solutions to the wave equation take the form:

$$ u(x,t) = A \cos(\omega t + kx) $$

where \( \omega \) and \( k \) relate to the wave speed and frequency.

### 5. Limits of Sequences

A sequence is a function whose domain is the set of natural numbers. The **limit of a sequence** represents the value that the terms of the sequence approach as the index increases indefinitely.

#### 5.1. Evaluating Limits of Sequences

A common technique to evaluate limits of rational sequences involves dividing the numerator and denominator by the highest power of \( n \). For example:

$$ \lim_{n \to \infty} \frac{n^2 + 3n}{2 n^2 - 2n} $$

Dividing by \( n^2 \):

$$ \lim_{n \to \infty} \frac{1 + \frac{3}{n}}{2 - \frac{2}{n}} = \frac{1 + 0}{2 - 0} = \frac{1}{2} $$

Similarly, for sequences involving cubic terms:

$$ \lim_{n \to \infty} \frac{(2n+3)^3}{n^3-1} $$

Expanding and simplifying yields the limit.

#### 5.2. The Squeeze Theorem

The **Squeeze Theorem** states that if a sequence \( a_n \) is squeezed between two other sequences that converge to the same limit, then \( a_n \) must also converge to that limit. This is particularly useful for oscillatory sequences like:

$$ \lim_{n \to \infty} \frac{\sin(n)}{n} $$

Since \( -1 \leq \sin(n) \leq 1 \), dividing by \( n \) and applying the squeeze theorem shows that the limit is \( 0 \).

#### 5.3. Exponential Sequences

An important limit is:

$$ a_n = \left(1 + \frac{1}{n}\right)^n $$

Taking the natural logarithm and applying L'HÃ´pital's Rule proves that:

$$ \lim_{n \to \infty} \left(1 + \frac{1}{n}\right)^n = e $$

### 1. Calculate:
#### (a) $$\displaystyle \lim_{n \to \infty} \frac{n^2 + 3n}{2 n^2 - 2n}$$

Let's break this down step by step.

1. Start by simplifying the expression:
   $$\frac{n^2 + 3n}{2 n^2 - 2n}$$

2. Divide both the numerator and the denominator by the highest power of $n$ in the expression, which is $n^2$. We do this to examine the behavior of the sequence as $n \to \infty$:
   $$\frac{\frac{n^2 + 3n}{n^2}}{\frac{2n^2 - 2n}{n^2}} = \frac{1 + \frac{3}{n}}{2 - \frac{2}{n}}$$

3. Now, as $n \to \infty$, both $\frac{3}{n}$ and $\frac{2}{n}$ approach $0$.

4. Therefore, the expression becomes:
   $$\frac{1 + 0}{2 - 0} = \frac{1}{2}$$

Hence, the limit is:
   $$\displaystyle \lim_{n \to \infty} \frac{n^2 + 3n}{2 n^2 - 2n} = \frac{1}{2}$$

#### (b) $$\displaystyle \lim_{n \to \infty} \frac{(2n+3)^3}{n^3-1}$$

1. Start by expanding the numerator:
   $$ (2n + 3)^3 = 8n^3 + 12n^2 + 18n + 27 $$

2. Rewrite the limit expression as:
   $$ \frac{8n^3 + 12n^2 + 18n + 27}{n^3 - 1} $$

3. Now, divide both the numerator and the denominator by $n^3$:
   $$ \frac{\frac{8n^3 + 12n^2 + 18n + 27}{n^3}}{\frac{n^3 - 1}{n^3}} = \frac{8 + \frac{12}{n} + \frac{18}{n^2} + \frac{27}{n^3}}{1 - \frac{1}{n^3}} $$

4. As $n \to \infty$, the terms $\frac{12}{n}$, $\frac{18}{n^2}$, $\frac{27}{n^3}$, and $\frac{1}{n^3}$ all approach $0$.

5. Therefore, the expression becomes:
   $$ \frac{8 + 0 + 0 + 0}{1 - 0} = 8 $$

Hence, the limit is:
   $$ \displaystyle \lim_{n \to \infty} \frac{(2n+3)^3}{n^3-1} = 8 $$

### 2. Prove using the squeeze theorem:  
$$\displaystyle \lim_{n \to \infty} \frac{\sin(n)}{n}$$

To prove this limit using the squeeze theorem, we will first observe that for all $n$, we know that:

$$ -1 \leq \sin(n) \leq 1 $$

Now divide through by $n$ (where $n > 0$):

$$ -\frac{1}{n} \leq \frac{\sin(n)}{n} \leq \frac{1}{n} $$

Next, observe that as $n \to \infty$, both $-\frac{1}{n}$ and $\frac{1}{n}$ approach $0$:

$$ \lim_{n \to \infty} -\frac{1}{n} = 0 \quad \text{and} \quad \lim_{n \to \infty} \frac{1}{n} = 0 $$

Since $\frac{\sin(n)}{n}$ is squeezed between two expressions that both approach $0$, by the squeeze theorem, we conclude that:

$$ \lim_{n \to \infty} \frac{\sin(n)}{n} = 0 $$

### 4. Find the limit of the sequence:
$$ a_n = \left( 1 + \frac{1}{n} \right)^n $$

This sequence is a well-known limit. We will prove that:

$$ \lim_{n \to \infty} \left( 1 + \frac{1}{n} \right)^n = e $$

1. Consider the sequence:
   $$ a_n = \left( 1 + \frac{1}{n} \right)^n $$

2. As $n \to \infty$, this expression is the definition of the number $e$. To see this, recall that the limit definition of $e$ is:
   $$ e = \lim_{n \to \infty} \left( 1 + \frac{1}{n} \right)^n $$

Hence, the limit is:
   $$ \lim_{n \to \infty} \left( 1 + \frac{1}{n} \right)^n = e $$
