## Limits of Real Functions

In calculus, **limits** are fundamental in understanding how a function behaves as the input (often denoted as \( x \)) approaches a certain value. The limit helps us analyze the behavior of a function near specific points, including at infinity or at points where the function may not be defined.

### 1. **Concept of Limits**

The limit of a function \( f(x) \) as \( x \) approaches a value \( a \), denoted as:

$$
\lim_{x \to a} f(x)
$$

is the value that \( f(x) \) approaches as \( x \) gets closer to \( a \). This means that for any small positive value \( \epsilon \), there exists a small \( \delta \) such that when \( x \) is within a distance of \( \delta \) from \( a \), \( f(x) \) is within a distance of \( \epsilon \) from the limit.

### 2. **Limits at Infinity**

When a function approaches infinity, we are interested in understanding the behavior of the function as \( x \) becomes larger and larger. This can be expressed as:

$$
\lim_{x \to \infty} f(x)
$$

If the function approaches a finite value, we say it has a **horizontal asymptote**. If the function grows without bound, we say the limit is infinity.

Example:
$$
\lim_{x \to \infty} \frac{1}{x} = 0
$$
In this example, as \( x \) increases, the value of \( \frac{1}{x} \) becomes closer to 0, indicating that the function has a horizontal asymptote at \( y = 0 \).

### 3. **One-Sided Limits**

One-sided limits refer to the behavior of the function as \( x \) approaches a specific value from either the left (\( x \to a^- \)) or the right (\( x \to a^+ \)). The one-sided limits are important when dealing with functions that may behave differently depending on the direction from which \( x \) approaches the point.

Example:
$$
\lim_{x \to 0^+} \frac{1}{x} = +\infty
$$

### 4. **Infinite Limits**

An infinite limit occurs when the function increases or decreases without bound as \( x \) approaches a certain value. In other words, as \( x \) approaches the limit point, the function grows without bound (goes to \( +\infty \) or \( -\infty \)).

Example:
$$
\lim_{x \to 0} \frac{1}{x^2} = +\infty
$$
As \( x \to 0 \), \( \frac{1}{x^2} \) increases indefinitely.

### 5. **L'Hopital's Rule**

L'Hopital's Rule is a powerful tool for evaluating limits of indeterminate forms, such as \( \frac{0}{0} \) or \( \frac{\infty}{\infty} \). The rule states that if the limit results in an indeterminate form, we can differentiate the numerator and the denominator separately and then take the limit again.

Example:
$$
\lim_{x \to 0} \frac{\sin(x)}{x} = 1
$$
By applying L'Hopital's Rule, we differentiate the numerator and denominator:
$$
\frac{\cos(x)}{1} \quad \text{and taking the limit as} \quad x \to 0, \quad \frac{\cos(0)}{1} = 1
$$

### 6. **Asymptotes**

Asymptotes are lines that a graph approaches but never touches. There are two main types of asymptotes:

- **Vertical Asymptotes**: These occur when the function approaches infinity as \( x \) approaches a specific value \( a \). The function gets closer to \( \pm \infty \) but never actually reaches that value.
  $$ \lim_{x \to a} f(x) = \pm \infty $$

- **Horizontal Asymptotes**: These occur when the function approaches a constant value as \( x \) approaches \( \pm \infty \). The graph approaches a horizontal line but never crosses it.
  $$ \lim_{x \to \infty} f(x) = L $$

### 7. **Computing Limits**

To compute limits, we can use various techniques, such as:

- **Direct Substitution**: Substitute the value of \( x \) into the function (if the function is continuous).
- **Factorization**: Factor the function and cancel common terms.
- **Rationalization**: Multiply by the conjugate to simplify expressions with square roots.
- **Use of Special Limits**: Apply known standard limits like \( \lim_{x \to 0} \frac{\sin(x)}{x} = 1 \) or \( \lim_{x \to \infty} \frac{1}{x} = 0 \).

---

### 8. **Examples of Limits**

Here are a couple of examples to better understand limits:

#### Example 1: Horizontal Asymptote
$$ \lim_{x \to \infty} \frac{2x^2 + 3x}{x^2 + 1} $$

- Divide both the numerator and the denominator by \( x^2 \) to simplify:
  $$ \lim_{x \to \infty} \frac{2 + \frac{3}{x}}{1 + \frac{1}{x^2}} $$

- As \( x \to \infty \), both \( \frac{3}{x} \) and \( \frac{1}{x^2} \) approach 0, leaving us with:
  $$ \frac{2}{1} = 2 $$

So, the horizontal asymptote is \( y = 2 \).

#### Example 2: Vertical Asymptote
$$ \lim_{x \to 0} \frac{1}{x} $$

- As \( x \to 0 \), the function approaches infinity (positive or negative depending on the direction), indicating a vertical asymptote at \( x = 0 \).

#### $$\displaystyle\lim_{x \to \infty} \frac{x^3 + 2x^2}{x^4 - 3x^3}$$

1. Begin by analyzing the given expression:
   $$ \frac{x^3 + 2x^2}{x^4 - 3x^3} $$

2. Divide both the numerator and the denominator by $x^4$, the highest power of $x$ in the denominator:
   $$ \frac{\frac{x^3 + 2x^2}{x^4}}{\frac{x^4 - 3x^3}{x^4}} = \frac{\frac{1}{x} + \frac{2}{x^2}}{1 - \frac{3}{x}} $$

3. Now, as $x \to \infty$, both $\frac{1}{x}$, $\frac{2}{x^2}$, and $\frac{3}{x}$ approach $0$.

4. Therefore, the expression becomes:
   $$ \frac{0 + 0}{1 - 0} = 0 $$

Hence, the limit is:
   $$ \displaystyle \lim_{x \to \infty} \frac{x^3 + 2x^2}{x^4 - 3x^3} = 0 $$

### 2. Find:
#### $$\displaystyle \lim_{x \to 0} \frac{\sin(3x)}{2x+1}$$

1. Start by substituting $x = 0$ into the function:
   $$ \frac{\sin(3 \cdot 0)}{2(0) + 1} = \frac{\sin(0)}{1} = \frac{0}{1} = 0 $$

Therefore, the limit is:
   $$ \displaystyle \lim_{x \to 0} \frac{\sin(3x)}{2x+1} = 0 $$

### 4. Find the asymptotes of the function:
#### $$f(x) = \frac{x^2 - 1}{x^2 + 1}$$

1. **Vertical Asymptotes:**
   Vertical asymptotes occur when the denominator approaches $0$ and the numerator does not approach $0$. For $f(x)$:
   $$ x^2 + 1 = 0 $$
   There is no real solution for this equation, since $x^2 + 1 > 0$ for all real values of $x$. Hence, there are **no vertical asymptotes**.

2. **Horizontal Asymptotes:**
   To find the horizontal asymptote, examine the behavior as $x \to \infty$:
   $$ f(x) = \frac{x^2 - 1}{x^2 + 1} $$

   Dividing both the numerator and the denominator by $x^2$, we get:
   $$ f(x) = \frac{1 - \frac{1}{x^2}}{1 + \frac{1}{x^2}} $$

   As $x \to \infty$, both $\frac{1}{x^2} \to 0$, so the function approaches:
   $$ f(x) = \frac{1}{1} = 1 $$

Hence, the horizontal asymptote is:
   $$ y = 1 $$

---

#### $$g(x) = \frac{\sin(x)}{x^2+1}$$

1. **Vertical Asymptotes:**
   Vertical asymptotes occur when the denominator approaches $0$ and the numerator does not approach $0$. For $g(x)$:
   $$ x^2 + 1 = 0 $$
   Again, there is no real solution, as $x^2 + 1 > 0$ for all real values of $x$. Hence, there are **no vertical asymptotes**.

2. **Horizontal Asymptotes:**
   To find the horizontal asymptote, examine the behavior as $x \to \infty$:
   $$ g(x) = \frac{\sin(x)}{x^2 + 1} $$

   As $x \to \infty$, the numerator $\sin(x)$ oscillates between $-1$ and $1$, while the denominator $x^2 + 1$ grows without bound. Therefore, the function approaches $0$ as $x \to \infty$.

Hence, the horizontal asymptote is:
   $$ y = 0 $$
